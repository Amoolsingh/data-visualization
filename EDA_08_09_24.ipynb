```python
#student
# DATA CLEANING...DROP CERTAIN COLUMNS
#axis=1,particular column(refers to vertical axis or columns)
# axis=0 for particular row
student=df.drop(['Attendance'],axis=1)
student
# we are able to drop attendance column
# drop labscore and final exam
```

```python
student=df.drop(['Lab Score','Final Exam'],axis=1)
student
#df.drop([])
```

```python
# How to find corelation of a function
# Numeric_only= True,enables only only numeric values
corelation=student.corr(numeric_only=True)
```

```python
corelation
# CORELATION IS A STATISTICAL MEASURE TO EXPRESS HOW WELL TWO VARIABLES ARE LINEARLY CONNECTED
#POSITIVE VALUE=pOSITIVE ASSOCIATION
# nEGATIVE VALUE=WE HAVE NEGATIVE ASSOCIATION
```

```python
# The corelation coefficient is a value that indicates the strength of the relationship between variables
# The coefficient  takes any value from -1 to 1,#rxy find the corelation(VALUE LIES BETWEEN 0 TO 1)
#-1 Perfect Negative correlation
#0, No corelation
#1 Perfect Positive corelation
```

```python
#heatmap  
#annot= True,The annot parameter in the seaborn.heatmap function displays the data values in each cell of 
# heatmap
sns.heatmap(corelation,annot= True)
# Heatmap is a two dimensional representation of data in which values are represented by colors
# A simple heatmap it provides you the data visualization summary of your corelation
```
