```python
dataset_table1=pd.crosstab(df['sex'],df['smoker'])
```

```python
dataset_table1
```

```python
observed_values=dataset_table1.values
```

```python
observed_values
```

```python
import seaborn as sns
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency
```

```python
val=stats.chi2_contingency(dataset_table1)
val
```

```python
Expected_Values=val[3]
Expected_Values
```

```python
no_of_rows=len(dataset_table1.iloc[0:2,0])
no_of_columns=len(dataset_table1.iloc[0,0:2])
```

```python
ddof=(no_of_rows-1)*(no_of_columns-1)
alpha=0.05
ddof
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(observed_values,Expected_Values)])
chi_square
```

```python
chi_square_statistic=chi_square[0]+chi_square[1]
chi_square_statistic
```

```python
critical_value=chi2.ppf(1-alpha,df=ddof)
critical_value
```

```python
if chi_square_statistic>=critical_value:
    print("Reject Ho(Null Hypothesis"),
else:
    print("Retain Ho(Null Hypothesis")
```
