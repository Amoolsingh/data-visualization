```python
from sklearn.datasets import make_classification
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
# number of sample by default its 100 the number feature is 20bydefault , number of informative is 2bydefaults, number of redundant is 2bydefaults
#number of classes is 2 bydefaults number of cluster is 2 by deaults 
# flip_y in make classification default value is 0.01 meaning 10%  whoses classes assigned randomly usually 
# noice or the outlier make the task harder
#
# step 2
from sklearn.datasets import make_classification
x,y= make_classification(
n_samples=100,
n_features=1,
n_classes=2,
n_clusters_per_class=1,
flip_y=0.03,
n_informative=1,
n_redundant=0)
```

```python
# step 4 spliting machine learning model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred=log_reg.predict(x_train) # predicting training model
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,y_pred)
```

```python
confusion_matrix(y_train,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_pred))
```
